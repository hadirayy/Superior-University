```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
from nltk.chat.util import Chat, reflections
```

```python
pairs = [
    (r"(hi|hello|hey|helo)", [
        "Hi there! Welcome to Superior University admission help.",
        "Hello! How can I assist you with admissions today?"
    ]),

    (r"how are you\??", [
        "I'm just a bot, but I'm doing great! How about you?"
    ]),

    (r"what is your purpose\??", [
        "I am here to assist you with any questions or tasks you have."
    ]),

    (r"(bye|goodbye|see you|exit|quit)", [
        "Goodbye! Have a great day!",
        "See you later!"
    ]),
    (r"what .*programs.*", [
        "Superior University offers BS, MS, and PhD programs across Computer Science, Business, Engineering, and Social Sciences."
    ]),

    (r"how can i apply.*", [
        "You can apply online through the official Superior University website: https://www.superior.edu.pk (check the 'Admissions' section)."
    ]),

    (r"(last date|deadline).*admission", [
        "Admission deadlines vary by semester (e.g., Fall usually in Aug, Spring usually in Feb). Please check the official website for exact dates."
    ]),

    (r"(admission requirements|requirements for admission|what are the requirements)", [
        "Requirements depend on the program. For BS: usually at least 45% in Intermediate. For MS: a relevant BS with minimum CGPA ~2.0. Check the program page for exact criteria."
    ]),

    (r"scholarship", [
        "Yes — Superior University offers merit-based, need-based, and sports scholarships for eligible students."
    ]),

    (r"(where .*superior university|location|address)", [
        "Superior University's main campus is at 17-KM Raiwind Road, Lahore, Pakistan."
    ]),

    (r"(contact|phone|email).*admission", [
        "You can contact the Admissions Office at (042) 38103777 or email info@superior.edu.pk for queries."
    ]),

    (r"hostel", [
        "Yes, the university provides separate hostels for male and female students with basic amenities."
    ]),

    (r"(admission test|entry test|interview)", [
        "Some programs require an entry test or interview. After applying, the admission office will notify shortlisted candidates via email/SMS."
    ]),

    (r"(online|distance|blended).*class", [
        "Yes — some programs offer online or blended learning options via the university's LMS."
    ]),
    (r"(.*)", [
        "I'm sorry, I don't understand that. You can ask about admissions, programs, deadlines, scholarships, or contact details. Or type 'sentiment' to analyze text sentiment."
    ])
]
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment(text):
    score = sia.polarity_scores(text)
    comp= score['compound']
    if comp >=0.05:
        return "Positive Sentence"
    elif comp <=-0.05:
        return "Negative Sentence"
    else:
        return "Neutral Sentence"
chatbot=Chat(pairs,reflections)
```

```python

def Jarvis():
    print("Jarvis: Hello — I'm Jarvis. (type 'bye' or 'exit' to quit, type 'sentiment' to analyze a sentence)")
    while True:
        user_input = input("You: ").strip()
        if not user_input:
            continue
        if user_input.lower() in ('bye', 'exit', 'quit'):
            print("Jarvis: Goodbye! Have a good day.")
            break
        if 'sentiment' in user_input.lower():
            sentiment_input = input("Sentiment Text: ").strip()
            if sentiment_input:
                print("Jarvis: Sentiment Analysis Result -", analyze_sentiment(sentiment_input))
            else:
                print("Jarvis: No text provided for sentiment analysis.")
            continue
        response = chatbot.respond(user_input)
        if response:
            print("Jarvis:", response)
        else:
            print("Jarvis: I'm sorry, I don't understand that. Try asking about admissions or type 'sentiment'.")

if __name__ == "__main__":
    Jarvis()
```
